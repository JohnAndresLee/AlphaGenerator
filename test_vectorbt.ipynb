```python
import vectorbt as vbt
import pandas as pd
import numpy as np
from numba import njit
```

```python
factor_data = pd.read_csv('./factors/Div(WMA(WMA($close,10),10),$close).csv')
factor_data = factor_data[factor_data['instrument']=='BTCUSDT_15M_DATA']
factor_data
```

```python
label_data = pd.read_csv('./my_data/basic/BTCUSDT_15m_data.csv')
label_data['label'] = (label_data['Close'].shift(-1)/label_data['Close'])-1
label_data = label_data[['Open_time', 'label']]
label_data.rename(columns={'Open_time':'datetime'}, inplace=True)
label_data
```

```python
data = pd.merge(factor_data, label_data, on='datetime', how='left')
data['datetime'] = pd.to_datetime(data['datetime'])
data
```

```python
tmp = data[(data['datetime'].dt.month==6)&(data['datetime'].dt.day==3)]
tmp['Div(WMA(WMA($close,10),10),$close)'].corr(tmp['label'])
```
